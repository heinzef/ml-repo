```python
from sklearn.datasets import fetch_openml
mnist = fetch_openml('mnist_784', version = 1)
```

# View Data

```python
mnist['data']
```

# View Real Written Numbers (Targets)


```python
mnist['target']
```

# Split Data into Training Dataset and Testing Dataset

```python
split_after = 60000

train_data = mnist['data'].iloc[0:split_after]
train_real_numbers = mnist['target'].iloc[0:split_after]

test_data = mnist['data'].iloc[split_after:]
test_real_numbers = mnist['target'].iloc[split_after:]
```

# View Training and Test Dataset

```python
train_data
```

```python
test_data
```

# Implement Function for kNN

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, confusion_matrix
def use_kNN(neighbors = 3):
    neigh = KNeighborsClassifier(n_neighbors=neighbors)
    neigh.fit(train_data, train_real_numbers)
    prediction = neigh.predict(test_data)
    print(classification_report(test_real_numbers, prediction))
    print(confusion_matrix(test_real_numbers, prediction))
```

# Test kNN with neighbors = 3 and time it

```python
%%time
use_kNN()
```

# Test kNN with neighbors = 1 and time it

```python
%%time
use_kNN(1)
```
